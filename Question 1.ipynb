```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
homicides = pd.read_csv("API_VC.IHR.PSRC.P5_DS2_en_csv_v2_4696515.csv", skiprows=3)

homicides.head()
```

```python
homicides.set_index('Country Code', inplace=True)

homicides.index
```

```python
gdp_pc = pd.read_csv("API_NY.GDP.PCAP.CD_DS2_en_csv_v2_4701206.csv",skiprows=3)
gdp_pc.set_index('Country Code', inplace=True)

pop = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_4701113.csv", skiprows=3)
pop.set_index('Country Code', inplace=True)
 
gdp_pc.head()
```

```python
pop.head()
```

```python
"""Load the country level meta data from 
'Metadata_Country_API_VC.IHR.PSRC.P5_DS2_en_csv_v2_4696515.csv'
Index all the tables by country code. 
4. Calculate means for the period 2010 to 2021 for the three variables (homicides, 
wealth, and population)  
5. For the purposes of this exercise, we focus on ‘higher income’ countries, filter your 
data to extract data for these countries. """
```

```python
meta_data = pd.read_csv('Metadata_Country_API_VC.IHR.PSRC.P5_DS2_en_csv_v2_4696515.csv')
meta_data.set_index('Country Code', inplace=True)

meta_data.head()
```

```python
homicides_mean = homicides.loc[:, '2010':'2021'].mean(axis=1)
gdp_pc_mean = gdp_pc.loc[:, '2010':'2021'].mean(axis=1)
population_mean = pop.loc[:, '2010':'2021'].mean(axis=1)
```

```python
homicides_mean
```

```python
higher_income = meta_data[meta_data['IncomeGroup']== 'High income'].index
homicides_higher_income = homicides.loc[higher_income]
gdp_pc_higher_income = gdp_pc.loc[higher_income]
population_higher_income = pop.loc[higher_income]
```

```python
combined_df = pd.DataFrame({
    'Homicides': homicides_mean,
    'GDP per Capita': gdp_pc_mean
})

# Join the metadata to get the region
combined_df = combined_df.join(meta_data[['Region']])

# Create the relplot, using 'Region' for the hue
sns.relplot(
    data=combined_df, 
    x='GDP per Capita', 
    y='Homicides', 
    hue='Region',
    kind='scatter',
    height=6, aspect=1.5
)

# Set the title and labels
plt.title('Relationship between Homicides and Wealth by Region')
plt.xlabel('GDP per Capita')
plt.ylabel('Homicides per 100,000 People')

# Save the plot to a PDF file
plt.savefig("homicides_vs_wealth_by_region.pdf", format="pdf")

# Show the plot
plt.show()
```

```python



# Combine the data into a single DataFrame
combined_df = pd.DataFrame({
    'Homicides': homicides_mean,
    'GDP per Capita': gdp_pc_mean
})

# Join the metadata to get the region
combined_df = combined_df.join(meta_data[['Region']])

# Apply log transformation to both 'Homicides' and 'GDP per Capita', avoiding log of zero values
combined_df['Log Homicides'] = np.log(combined_df['Homicides'] + 0.01)
combined_df['Log GDP per Capita'] = np.log(combined_df['GDP per Capita'] + 0.01)

# Create the relplot using the log-transformed data
sns.relplot(
    data=combined_df, 
    x='Log GDP per Capita', 
    y='Log Homicides', 
    hue='Region',
    kind='scatter',
    height=6, aspect=1.5
)

# Set the title and labels
plt.title('Log-Log Relationship between Homicides and Wealth by Region')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Log Homicides per 100,000 People')

# Save the plot to a PDF file
plt.savefig("log_homicides_vs_log_wealth_by_region.pdf", format="pdf")

# Show the plot
plt.show()
```

```python
# Plot the relationship with a regression line using lmplot on the log-transformed data
sns.lmplot(
    data=combined_df, 
    x='Log GDP per Capita', 
    y='Log Homicides', 
    hue='Region',
    height=6, aspect=1.5,
    scatter_kws={'alpha':0.5},  # Make the scatter points slightly transparent
    line_kws={'color': 'black'}  # Make the regression line black
)

# Set the title and labels
plt.title('Log-Log Relationship between Homicides and Wealth (with Linear Fit)')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Log Homicides per 100,000 People')

# Save to PDF
plt.savefig("log_log_lmplot_homicides_wealth.pdf", format="pdf")

# Show the plot
plt.show()
```

```python
# Rank countries by population and wealth
combined_df['Population'] = population_mean
combined_df['Wealth'] = gdp_pc_mean

# Rank by population and wealth
combined_df['Population Rank'] = combined_df['Population'].rank(ascending=False)
combined_df['Wealth Rank'] = combined_df['Wealth'].rank(ascending=False)

# Filter countries in the top half for both population and wealth
filtered_df = combined_df[
    (combined_df['Population Rank'] <= combined_df['Population Rank'].median()) &
    (combined_df['Wealth Rank'] <= combined_df['Wealth Rank'].median())
]
```

```python
# Plot Homicides vs Wealth for large, wealthy countries
sns.relplot(
    data=filtered_df, 
    x='Wealth', 
    y='Homicides', 
    hue='Region',
    kind='scatter',
    height=6, aspect=1.5
)

# Set the title and labels
plt.title('Homicides vs Wealth for Large, Wealthy Countries')
plt.xlabel('GDP per Capita')
plt.ylabel('Homicides per 100,000 People')

# Save the plot to a PDF
plt.savefig("homicides_vs_wealth_large_wealthy_countries.pdf", format="pdf")

# Show the plot
plt.show()
```
